```python
# - 1.8 Encuentra cuántas filas tienen valores nulos en price y rellena esos valores con el promedio de la columna.
# nulos_price = vinos_df['price'].isnull().sum()
# print(f'Número de nulls ANTES de rellenar: {nulos_price}')


# filtro = vinos_df['price'].isnull()
# promedio =vinos_df['price'].mean()

# vinos_df.loc[filtro, 'price'] = promedio

# nulos_price = vinos_df['price'].isnull().sum()
# print(f'Número de nulls DESPUES de rellenar: {nulos_price}')


# Método más directo
nulos_price = vinos_df['price'].isnull().sum()
print(f'Número de nulls ANTES de rellenar: {nulos_price}')

vinos_df['price'] = vinos_df['price'].fillna(vinos_df['price'].mean())

nulos_price = vinos_df['price'].isnull().sum()
print(f'Número de nulls DESPUES de rellenar: {nulos_price}')
```

```python
# Calcular el promedio de la columna 'price'
average_price = vinos_df['price'].mean()

# Filtrar vinos con 'price' mayor que el promedio
wines_above_average = vinos_df[vinos_df['price'] > average_price]

# Contar cuántos vinos cumplen la condición
count = wines_above_average.shape[0]

# Calcular el porcentaje
percentage = (count / vinos_df['price'].notnull().sum()) * 100

# Mostrar resultados
print(f"Cantidad de vinos con un precio mayor al promedio: {count}")
print(f"Porcentaje de vinos con un precio mayor al promedio: {percentage:.2f}%")

# Comprobación
print(f"Promedio del precio: {average_price}")
print(vinos_df['price'].describe())
```

```python
# AVANZADO
# - 2.1 Rellena valores nulos en region_1 y reemplázalos por "unknown"

print(f'Antes de rellenar: {vinos_df["region_1"].isnull().sum()}')

# Agregar 'unknown' como una categoría válida - Por usar tipo de dato CATEGORY  es necesario agregar una categoría nueva y no la cadena 'unknown' directamente.
vinos_df['region_1'] = vinos_df['region_1'].cat.add_categories(['unknown'])

# Reemplazar valores nulos con 'unknown'
vinos_df['region_1'] = vinos_df['region_1'].fillna('unknown')

# Verificar que no haya valores nulos
print(f'Después de rellenar: {vinos_df["region_1"].isnull().sum()}')  # Esto debería devolver 0

```

```python
# - 2.2 ¿Cuantos vinos tienen points mayor o igual a 90 y un precio menor o igual a 20? Usa un filtro lógico para calcularlo


# Crear el filtro lógico
filtro = vinos_df[(vinos_df['points'] >= 90) & (vinos_df['price'] <= 20)]

# Contar cuántos vinos cumplen la condición
count = filtro.shape[0]

print(f"Cantidad de vinos con puntos >= 90 y precio <= 20: {count}")

# Porcentaje de vinos con puntos >= 90 y precio <= 20
total_vinos = vinos_df.shape[0]
percentage = (count / total_vinos) * 100
print(f"Porcentaje de vinos con puntos >= 90 y precio <= 20: {percentage:.2f}%")

```

```python
# Otra forma

# Filtrar con .between() para puntos y precio
count = vinos_df[(vinos_df['points'].between(90, 100)) & (vinos_df['price'].between(0, 20))].shape[0]

print(f"Cantidad de vinos con puntos >= 90 y precio <= 20: {count}")
```

```python
# AVANZADO 2
# - 3.1 ¿Qué vino es la "mejor oferta"? (puntos/precio)(puedes usar .loc y la función idmax())
# Crear una nueva columna para puntos/precio
vinos_df['points_per_price'] = vinos_df['points'] / vinos_df['price']

# Encontrar el índice del vino con la mejor oferta
best_offer_index = vinos_df['points_per_price'].idxmax()

# Usar .loc para mostrar el vino con la mejor oferta
best_offer = vinos_df.loc[best_offer_index]
print("Mejor oferta:")
print(best_offer[['title', 'points', 'price', 'points_per_price']])
```

```python
# - 3.2 Calcula la media de puntos para cada país con groupby
#   - 3.2.1 Utiliza la función sort_values para ordenar el resultado por orden descendente de puntuación media.
# Calcular la media de puntos por país
mean_points_by_country = vinos_df.groupby('country')['points'].mean()

# Ordenar por puntuación media descendente
sorted_mean_points = mean_points_by_country.sort_values(ascending=False)

print("Media de puntos por país (ordenada):")
print(sorted_mean_points)
```

```python
# - 3.3 Encuentra la provincia que tiene el vino más caro (price) en cada país. Muestra el país, la provincia y el precio más alto. (puedes usar loc + groupby)
# Crear un DataFrame con el precio máximo por país
max_price_by_country = vinos_df.groupby('country', observed=True)['price'].max().reset_index()

# Combinar con el DataFrame original para obtener las provincias correspondientes
province_max_price = vinos_df.merge(max_price_by_country, on=['country', 'price'], how='inner')[['country', 'province', 'price']]

print("Provincia con el vino más caro en cada país:")
print(province_max_price)
```
