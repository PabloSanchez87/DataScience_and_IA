## Distribución de Ubicaciones de las Empresas (agrupando las menores como 'Otros')

```python
location_counts = salaries_df['company_location'].value_counts()

# Establecer el umbral para agrupar los países menos comunes como "Otros"
threshold = 10
grouped_locations = location_counts.copy()
grouped_locations[location_counts < threshold] = 0  # Asignar 0 a países con menos datos
grouped_locations["Others"] = location_counts[location_counts < threshold].sum()  # Sumar los datos de países menores


# Filtrar solo las ubicaciones con datos
grouped_locations = grouped_locations[grouped_locations > 0]


# Crear el gráfico
plt.figure(figsize=(12, 6))
grouped_locations.sort_values(ascending=False).plot(kind='bar')
plt.title("Distribución de Ubicaciones de las Empresas (agrupando las menores como 'Otros')")
plt.xlabel("Ubicación")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Distribución de Salarios (en USD) por Ubicación

```python
# Crear el gráfico de boxplot comparando la ubicación con el salario en USD
plt.figure(figsize=(14, 8))
sns.boxplot(data=salaries_df, x='company_location', y='salary_in_usd')

# Personalizar el gráfico
plt.title("Distribución de Salarios (en USD) por Ubicación", fontsize=16)
plt.xlabel("Ubicación", fontsize=12)
plt.ylabel("Salario en USD", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Muestra el gráfico
plt.show()
```

## Distribución de Salarios (en USD) por Ubicación (Agrupado)

```python
# Definir un umbral para agrupar países con pocas muestras como "Otros"
threshold = 10
country_counts = salaries_df['company_location'].value_counts()
salaries_df['grouped_location'] = salaries_df['company_location'].apply(
    lambda x: x if country_counts[x] >= threshold else 'Others'
)

# Crear el gráfico de boxplot con la columna agrupada
plt.figure(figsize=(14, 8))
sns.boxplot(data=salaries_df, x='grouped_location', y='salary_in_usd')

# Personalizar el gráfico
plt.title("Distribución de Salarios (en USD) por Ubicación (Agrupado)", fontsize=16)
plt.xlabel("Ubicación", fontsize=12)
plt.ylabel("Salario en USD", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Muestra el gráfico
plt.show()
```

# Evolución del Salario Promedio por Año

```python
import matplotlib.pyplot as plt

# Calcular el salario promedio por año
average_salaries = salaries_df.groupby('work_year')['salary_in_usd'].mean()

# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))
average_salaries.plot(kind='line', marker='o')
plt.title("Evolución del Salario Promedio por Año", fontsize=16)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Salario Promedio (en USD)", fontsize=12)
plt.grid()
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Crear el gráfico de boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=salaries_df, x='work_year', y='salary_in_usd')

# Personalizar el gráfico
plt.title("Distribución de Salarios por Año", fontsize=16)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Salario (en USD)", fontsize=12)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

# Gráfico de barras apiladas por año y nivel de experiencia

```python
# Agrupar datos por año y nivel de experiencia, calculando el salario promedio
experience_salary = salaries_df.groupby(['work_year', 'experience_level'])['salary_in_usd'].mean().unstack()

# Crear el gráfico de barras apiladas
experience_salary.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title("Salarios Promedios por Año y Nivel de Experiencia", fontsize=16)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Salario Promedio (en USD)", fontsize=12)
plt.legend(title="Nivel de Experiencia")
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

# Salario Promedio por Año y Nivel de Experiencia

```python
# Crear una tabla pivote para el heatmap
heatmap_data = salaries_df.pivot_table(
    index='work_year', columns='experience_level', values='salary_in_usd', aggfunc='mean'
)

# Crear el heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="coolwarm", linewidths=.5)
plt.title("Salario Promedio por Año y Nivel de Experiencia", fontsize=16)
plt.xlabel("Nivel de Experiencia", fontsize=12)
plt.ylabel("Año", fontsize=12)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

# Evolución del Salario Promedio por Ubicación

```python

# Agrupar los datos por año y localización, calculando el salario promedio
location_salary = salaries_df.groupby(['work_year', 'company_location'])['salary_in_usd'].mean().unstack()

# Crear el gráfico de líneas múltiples
plt.figure(figsize=(14, 8))
location_salary.plot(ax=plt.gca(), marker='o', alpha=0.7)

# Personalizar el gráfico
plt.title("Evolución del Salario Promedio por Ubicación", fontsize=16)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Salario Promedio (en USD)", fontsize=12)
plt.legend(title="Ubicación", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

# Evolución del Salario Promedio por Principales Ubicaciones

```python
# Seleccionar ubicaciones con más datos
top_locations = salaries_df['company_location'].value_counts().nlargest(10).index
filtered_df = salaries_df[salaries_df['company_location'].isin(top_locations)]

# Repetir el análisis con el DataFrame filtrado
location_salary_filtered = filtered_df.groupby(['work_year', 'company_location'])['salary_in_usd'].mean().unstack()

# Crear el gráfico
plt.figure(figsize=(10, 8))
location_salary_filtered.plot(ax=plt.gca(), marker='o', alpha=0.7)
plt.title("Evolución del Salario Promedio por Principales Ubicaciones", fontsize=16)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Salario Promedio (en USD)", fontsize=12)
plt.legend(title="Ubicación", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Crear la gráfica de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(salaries_df['salary'], salaries_df['salary_in_usd'], alpha=0.6)

# Personalizar el gráfico
plt.title("Comparación de Salarios en Moneda Local vs Salarios en USD", fontsize=16)
plt.xlabel("Salario en Moneda Local", fontsize=12)
plt.ylabel("Salario en USD", fontsize=12)
plt.grid(alpha=0.4)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

# Comparación de Salarios en Moneda Local vs Salarios en USD (por Año)

```python
# Crear la gráfica de dispersión considerando los años
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=salaries_df,
    x='salary', 
    y='salary_in_usd', 
    hue='work_year',  # Diferenciar por años usando colores
    palette='viridis', 
    alpha=0.7
)

# Personalizar el gráfico
plt.title("Comparación de Salarios en Moneda Local vs Salarios en USD (por Año)", fontsize=16)
plt.xlabel("Salario en Moneda Local", fontsize=12)
plt.ylabel("Salario en USD", fontsize=12)
plt.legend(title="Año", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(alpha=0.4)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
