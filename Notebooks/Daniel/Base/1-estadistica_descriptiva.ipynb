## DistribuciÃ³n de Ubicaciones de las Empresas (agrupando las menores como 'Otros')

```python
location_counts = salaries_df['company_location'].value_counts()

# Establecer el umbral para agrupar los paÃ­ses menos comunes como "Otros"
threshold = 10
grouped_locations = location_counts.copy()
grouped_locations[location_counts < threshold] = 0  # Asignar 0 a paÃ­ses con menos datos
grouped_locations["Others"] = location_counts[location_counts < threshold].sum()  # Sumar los datos de paÃ­ses menores


# Filtrar solo las ubicaciones con datos
grouped_locations = grouped_locations[grouped_locations > 0]


# Crear el grÃ¡fico
plt.figure(figsize=(12, 6))
grouped_locations.sort_values(ascending=False).plot(kind='bar')
plt.title("DistribuciÃ³n de Ubicaciones de las Empresas (agrupando las menores como 'Otros')")
plt.xlabel("UbicaciÃ³n")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## DistribuciÃ³n de Salarios (en USD) por UbicaciÃ³n

```python
# Crear el grÃ¡fico de boxplot comparando la ubicaciÃ³n con el salario en USD
plt.figure(figsize=(14, 8))
sns.boxplot(data=salaries_df, x='company_location', y='salary_in_usd')

# Personalizar el grÃ¡fico
plt.title("DistribuciÃ³n de Salarios (en USD) por UbicaciÃ³n", fontsize=16)
plt.xlabel("UbicaciÃ³n", fontsize=12)
plt.ylabel("Salario en USD", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Muestra el grÃ¡fico
plt.show()
```

## DistribuciÃ³n de Salarios (en USD) por UbicaciÃ³n (Agrupado)

```python
# Definir un umbral para agrupar paÃ­ses con pocas muestras como "Otros"
threshold = 10
country_counts = salaries_df['company_location'].value_counts()
salaries_df['grouped_location'] = salaries_df['company_location'].apply(
    lambda x: x if country_counts[x] >= threshold else 'Others'
)

# Crear el grÃ¡fico de boxplot con la columna agrupada
plt.figure(figsize=(14, 8))
sns.boxplot(data=salaries_df, x='grouped_location', y='salary_in_usd')

# Personalizar el grÃ¡fico
plt.title("DistribuciÃ³n de Salarios (en USD) por UbicaciÃ³n (Agrupado)", fontsize=16)
plt.xlabel("UbicaciÃ³n", fontsize=12)
plt.ylabel("Salario en USD", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Muestra el grÃ¡fico
plt.show()
```

# EvoluciÃ³n del Salario Promedio por AÃ±o

```python
import matplotlib.pyplot as plt

# Calcular el salario promedio por aÃ±o
average_salaries = salaries_df.groupby('work_year')['salary_in_usd'].mean()

# Crear el grÃ¡fico de lÃ­neas
plt.figure(figsize=(10, 6))
average_salaries.plot(kind='line', marker='o')
plt.title("EvoluciÃ³n del Salario Promedio por AÃ±o", fontsize=16)
plt.xlabel("AÃ±o", fontsize=12)
plt.ylabel("Salario Promedio (en USD)", fontsize=12)
plt.grid()
plt.tight_layout()

# Mostrar el grÃ¡fico
plt.show()
```

```python
# Crear el grÃ¡fico de boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=salaries_df, x='work_year', y='salary_in_usd')

# Personalizar el grÃ¡fico
plt.title("DistribuciÃ³n de Salarios por AÃ±o", fontsize=16)
plt.xlabel("AÃ±o", fontsize=12)
plt.ylabel("Salario (en USD)", fontsize=12)
plt.tight_layout()

# Mostrar el grÃ¡fico
plt.show()
```

# GrÃ¡fico de barras apiladas por aÃ±o y nivel de experiencia

```python
# Agrupar datos por aÃ±o y nivel de experiencia, calculando el salario promedio
experience_salary = salaries_df.groupby(['work_year', 'experience_level'])['salary_in_usd'].mean().unstack()

# Crear el grÃ¡fico de barras apiladas
experience_salary.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title("Salarios Promedios por AÃ±o y Nivel de Experiencia", fontsize=16)
plt.xlabel("AÃ±o", fontsize=12)
plt.ylabel("Salario Promedio (en USD)", fontsize=12)
plt.legend(title="Nivel de Experiencia")
plt.tight_layout()

# Mostrar el grÃ¡fico
plt.show()
```

# Salario Promedio por AÃ±o y Nivel de Experiencia

```python
# Crear una tabla pivote para el heatmap
heatmap_data = salaries_df.pivot_table(
    index='work_year', columns='experience_level', values='salary_in_usd', aggfunc='mean'
)

# Crear el heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="coolwarm", linewidths=.5)
plt.title("Salario Promedio por AÃ±o y Nivel de Experiencia", fontsize=16)
plt.xlabel("Nivel de Experiencia", fontsize=12)
plt.ylabel("AÃ±o", fontsize=12)
plt.tight_layout()

# Mostrar el grÃ¡fico
plt.show()
```

# EvoluciÃ³n del Salario Promedio por UbicaciÃ³n

```python

# Agrupar los datos por aÃ±o y localizaciÃ³n, calculando el salario promedio
location_salary = salaries_df.groupby(['work_year', 'company_location'])['salary_in_usd'].mean().unstack()

# Crear el grÃ¡fico de lÃ­neas mÃºltiples
plt.figure(figsize=(14, 8))
location_salary.plot(ax=plt.gca(), marker='o', alpha=0.7)

# Personalizar el grÃ¡fico
plt.title("EvoluciÃ³n del Salario Promedio por UbicaciÃ³n", fontsize=16)
plt.xlabel("AÃ±o", fontsize=12)
plt.ylabel("Salario Promedio (en USD)", fontsize=12)
plt.legend(title="UbicaciÃ³n", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.tight_layout()

# Mostrar el grÃ¡fico
plt.show()
```

# EvoluciÃ³n del Salario Promedio por Principales Ubicaciones

```python
# Seleccionar ubicaciones con mÃ¡s datos
top_locations = salaries_df['company_location'].value_counts().nlargest(10).index
filtered_df = salaries_df[salaries_df['company_location'].isin(top_locations)]

# Repetir el anÃ¡lisis con el DataFrame filtrado
location_salary_filtered = filtered_df.groupby(['work_year', 'company_location'])['salary_in_usd'].mean().unstack()

# Crear el grÃ¡fico
plt.figure(figsize=(10, 8))
location_salary_filtered.plot(ax=plt.gca(), marker='o', alpha=0.7)
plt.title("EvoluciÃ³n del Salario Promedio por Principales Ubicaciones", fontsize=16)
plt.xlabel("AÃ±o", fontsize=12)
plt.ylabel("Salario Promedio (en USD)", fontsize=12)
plt.legend(title="UbicaciÃ³n", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Crear la grÃ¡fica de dispersiÃ³n
plt.figure(figsize=(10, 6))
plt.scatter(salaries_df['salary'], salaries_df['salary_in_usd'], alpha=0.6)

# Personalizar el grÃ¡fico
plt.title("ComparaciÃ³n de Salarios en Moneda Local vs Salarios en USD", fontsize=16)
plt.xlabel("Salario en Moneda Local", fontsize=12)
plt.ylabel("Salario en USD", fontsize=12)
plt.grid(alpha=0.4)
plt.tight_layout()

# Mostrar el grÃ¡fico
plt.show()
```
